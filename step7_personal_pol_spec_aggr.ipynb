```python
import pandas as pd
import numpy as np
import os
from datetime import datetime
```

```python
path = r'D:\ChungyiTseng\Desktop\\'
exec_dt = datetime.strftime(datetime.now(), '%Y%m%d_%H%M%S')
```

```python
hate_df = pd.DataFrame()

for x in os.listdir(path):
    if x.startswith('hate_senti_fil'):
        print(x)
        hate_df = hate_df.append(pd.read_csv(path + x))
```

```python
hate_df.reset_index(drop=True, inplace=True)
```

```python
hate_df.shape
```

```python
hate_df.loc[hate_df.senti_sum > 0, 'pos_neg_flag'] = 'pos'
hate_df.loc[hate_df.senti_sum < 0, 'pos_neg_flag'] = 'neg'
```

```python
#hate_df.party.unique()

dpp_cond = hate_df.party.isin(['deep_dpp_kw', 'light_dpp_kw', 'tsai_kw'])
kmt_cond = hate_df.party.isin(['deep_kmt_kw', 'light_kmt_kw', 'han_kw'])
pos_cond = hate_df.pos_neg_flag =='pos'
neg_cond = hate_df.pos_neg_flag =='neg'

hate_df['party_aggr'] = hate_df.party + '_' + hate_df.pos_neg_flag
hate_df.loc[dpp_cond & pos_cond, 'party_aggr'] = 'dpp'
hate_df.loc[kmt_cond & pos_cond, 'party_aggr'] = 'kmt'
hate_df.loc[dpp_cond & neg_cond, 'party_aggr'] = 'kmt'
hate_df.loc[kmt_cond & neg_cond, 'party_aggr'] = 'dpp'
```

```python
hate_t1 = hate_df[~hate_df.party.isin(['undef'])].groupby(['author', 'party_aggr'], as_index=False).agg({'party':'count'})
```

```python
hate_t1['rnk'] = hate_t1.groupby('author')['party'].rank(method='dense', ascending=False)
```

```python
hate_t2 = hate_t1[ (hate_t1.rnk == 1) & (hate_t1.party != 1)]
```

```python
hate_t2.to_csv(path + 'aggr_author_{}.csv'.format(exec_dt))
```
